```python
from stereomolgraph.stereodescriptors import Stereo
from stereomolgraph import StereoMolGraph
from rdkit.Chem.rdmolfiles import MolFromSmiles
from rdkit.Chem.rdmolops import AddHs
from stereomolgraph.ipython import View2D
from stereomolgraph.experimental.canon import canon_atom_num
from stereomolgraph.experimental import generate_stereoisomers

View2D.show_h = False

# smi = "C[C@H]1C[C@H](C1)C"
smi = "CC1CC(C1)C"
rdmol = AddHs(MolFromSmiles(smi))
graph = StereoMolGraph.from_rdmol(rdmol, stereo_complete=False)
stereo_graphs = list(generate_stereoisomers(graph))
for stereo_graph in stereo_graphs:
    canon_id_mapping = canon_atom_num(stereo_graph)
    print(canon_id_mapping)
    stereo_graph = stereo_graph.relabel_atoms(canon_id_mapping)
    display(stereo_graph)
```
