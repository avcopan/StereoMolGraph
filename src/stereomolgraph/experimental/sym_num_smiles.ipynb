```python
import pandas as pd
from rdkit import Chem

from stereomolgraph import StereoMolGraph
from stereomolgraph.experimental.sym_num import topological_symmetry_number
```

```python
smiles = [
    "C1[C@H]2[C@H]3[C@@H]4[C@H]1[C@@H]4[C@@H]23",
    "[C@@H]12[C@H]3[C@H]4[C@@H]1[C@@H]5[C@H]4[C@H]3[C@H]25",
    "C1[C@H]2C[C@H]3C[C@@H]1C[C@@H](C2)C3",
    "F[C@H]1C[C@@H](F)C[C@@H](F)C1",
    "F[C@@H]1C[C@@H](F)C[C@@H](F)C1",
    "F[C@H]1[C@@H](F)C[C@@H](F)[C@@H](F)C1",
    "F[C@H]1[C@H](F)C[C@@H](F)[C@H](F)C1",
    "F[C@H]1[C@H](F)C[C@H](F)[C@@H](F)C1",
    "F[C@H]1[C@@H](F)C[C@H](F)[C@H](F)C1",
    "F[C@H]1[C@@H](F)[C@@H](F)[C@H](F)[C@H](F)[C@H]1F",
    "[C@@]12(CCC3)[C@@]3(CCC4)[C@@]4(CCC5)[C@]65[C@]1(CCC2)CCC6",
    "[C@@]12(CCC3)[C@@]3(CCC4)[C@@]4(CCC5)[C@]65[C@@]1(CCC2)CCC6",
    "[C@@]12(CCC3)[C@]3(CCC4)[C@@]4(CCC5)[C@]65[C@@]1(CCC2)CCC6",
    "C1=C2CC=C3[C@]24C(CC=C4C1)=CC3",
]
```

```python
rows = []
for smiles_string in smiles:
    rdmol = Chem.MolFromSmiles(smiles_string, sanitize=False)

    graph = StereoMolGraph.from_rdmol(rdmol, stereo_complete=True)
    rows.append(
        {
            "input_smiles": smiles_string,
            "top_sym_num": topological_symmetry_number(graph),
        }
    )

df_smiles_symmetry = pd.DataFrame(rows)
df_smiles_symmetry
```
