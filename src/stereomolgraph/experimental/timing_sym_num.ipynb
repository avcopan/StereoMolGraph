```python
from collections import deque

import rdkit
from rdkit import Chem

from stereomolgraph import StereoMolGraph
from stereomolgraph.algorithms.isomorphism import vf2pp_all_isomorphisms
```

```python
large_mol = (
    "CC(C)(C)C",
    "CC(C)(C)C(C)(C)C",
    "CC(C)(C)C(C)(C)C(C)(C)C",
    "CC(C)(C)C(C)(C)C(C)(C)C(C)(C)C",
)
# "CC(C)(C)C1=CC2=CC(CC3=CC(CC4=CC(C2)=CC(C(C)(C)C)=C4)=CC(C(C)(C)C)=C3)=C1"
# "CC(C)(C)C1=CC(CC2=CC(CC3=CC(C(C)(C)C)=CC(C4)=C3)=CC(C(C)(C)C)=C2)=CC(CC5=CC(C(C)(C)C)=CC(CC6=CC(CC7=CC(C(C)(C)C)=CC4=C7)=CC(C(C)(C)C)=C6)=C5)=C1"
```

```python
rdmol = Chem.AddHs(rdkit.Chem.MolFromSmiles(large_mol[3]))
smg = StereoMolGraph.from_rdmol(rdmol, stereo_complete=True)
```

```python
smg
```

```python
def num_mappings(mappings_generator) -> int:
    return deque(enumerate(mappings_generator, 1), maxlen=1)[0][0]
```

```python
num_mappings(vf2pp_all_isomorphisms(smg, smg, stereo=True))
```

```python
num_mappings(vf2pp_all_isomorphisms(smg, smg, stereo=False))
```
